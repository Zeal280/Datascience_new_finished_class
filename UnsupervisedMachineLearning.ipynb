```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('Mall_Customers.csv')
df.head()
```

```python
pd.plotting.andrews_curves(df, 'Gender')
```

```python
x = df.iloc[:,[3,4]].values
```

```python
wcss = []
for i in range (1,11):
    km = KMeans(n_clusters= i)
    km.fit(x)
    wcss.append(km.inertia_)
wcss
```

```python
plt.style.use('ggplot')
plt.plot(range(1,11),wcss)
plt.xlabel('RANGE')
plt.ylabel('WCSS')
plt.title('ELBOW METHOD')
plt.show
```

```python
km = KMeans(n_clusters=5)
ymeans = km.fit_predict(x)
ymeans
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(10,10))
plt.scatter(x[ymeans == 0,0], x[ymeans == 0,1], c='yellow', s=200)
plt.scatter(x[ymeans == 1,0], x[ymeans == 1,1], c='magenta', s=200)
plt.scatter(x[ymeans == 2,0], x[ymeans == 2,1], c='teal', s=200)
plt.scatter(x[ymeans == 3,0], x[ymeans == 3,1], c='navy', s=200)
plt.scatter(x[ymeans == 4,0], x[ymeans == 4,1], c='cyan', s=200)
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], s=200, c='black', label='cluster point')
plt.xlabel('ANNUAL INCOME')
plt.ylabel('SPENDING SCORE')
plt.title('GROUPING OF CUSTOMERS BASED ON ANNUAL INCOME AND SPENDING SCORE')
plt.legend(loc=[1,0])
plt.show()

```
